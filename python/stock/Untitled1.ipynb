```python
import fbprophet
```

```python
mdf = df[['Date','Open']].copy()
```

```python
mdf.columns=['ds','y']
```

```python
mdf.ds = pd.to_datetime(mdf.ds)
```

```python
year = '2019'
```

```python
m = fbprophet.Prophet(weekly_seasonality=False, daily_seasonality=False)
m.fit(mdf[mdf.ds < year])
```

```python
future = m.make_future_dataframe(350)
```

```python
forecast = m.predict(future)
```

```python
y1 = forecast[forecast.ds > year]
```
